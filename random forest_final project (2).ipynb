```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data = pd.read_excel('listings editted.xlsx')
data
```

```python
data = data.dropna(subset=['sentiment_overall'])
data.fillna(0, inplace=True)
data.fillna(0)
data
```

```python
X = data[['host_response_rate', 'host_acceptance_rate', 'host_is_superhost', 'host_total_listings_count', 'host_identity_verified', 'accommodates', 'bedrooms', 'beds','price','minimum_nights_avg_ntm','maximum_nights_avg_ntm','instant_bookable','calculated_host_listings_count','reviews_per_month']]
y_sent = data['sentiment_overall']
X_train, X_test, y_sent_train, y_sent_test = train_test_split(X, y_sent, test_size = 0.2)
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr_sent = regr.fit(X_train, y_sent_train)
regr_sent.feature_importances_
y_sent_test_pred = regr_sent.predict(X_test)
pd.DataFrame(list(X),list(regr_sent.feature_importances_))
```

```python
mse_sent_rf = mean_squared_error(y_sent_test_pred, y_sent_test)
mse_sent_rf
```

```python
r2_sent_rf = r2_score(y_sent_test,y_sent_test_pred)
r2_sent_rf
```

```python
y_rating = data['review_scores_rating']
X_train, X_test, y_rating_train, y_rating_test = train_test_split(X, y_rating, test_size = 0.2)
regr_rating = regr.fit(X_train, y_rating_train)
regr_rating.feature_importances_
y_rating_test_pred = regr_rating.predict(X_test)
pd.DataFrame(list(X),list(regr_rating.feature_importances_))
```

```python
mse_rating_rf = mean_squared_error(y_rating_test_pred, y_rating_test)
mse_rating_rf
```

```python
r2_rating_rf = r2_score(y_rating_test,y_rating_test_pred)
r2_rating_rf
```
